```python
# 使用query方法进行条件查询
df_filtered = df1.query('close > 500')
df_filtered
```

```python
# 计算涨跌幅
df1['涨跌幅'] = (df1['close'] - df1['open']) / df1['open'] * 100

# 使用query方法进行多条件查询
# 查询收盘价大于50且涨跌幅大于2%的数据
df_filtered2 = df1.query('close > 50 and 涨跌幅 > 2')

# 显示结果
df_filtered2
```

```python
# 导入matplotlib库
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False

# 设置图形大小
plt.figure(figsize=(12, 6))

# 绘制涨跌幅折线图
plt.plot(df1['day'], df1['涨跌幅'], linewidth=1)

# 设置标题和轴标签
plt.title('股票涨跌幅走势图')
plt.xlabel('日期')
plt.ylabel('涨跌幅(%)')

# 旋转x轴标签以防重叠
plt.xticks(rotation=45)

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.7)

# 显示图形
plt.tight_layout()  # 自动调整布局
plt.show()
```

```python
# 生成20天的K线图
df1.plot(kind='line', x='day', y='close', figsize=(12, 6))
plt.show()
```
